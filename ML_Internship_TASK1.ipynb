**Task 1: Linear Regression**

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/content/Salary_dataset.csv')

plt.figure(figsize=(10, 6))
plt.scatter(data['YearsExperience'], data['Salary'], color='blue', alpha=0.5)

plt.title('Scatter Diagram')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')

plt.grid()
plt.show()
```

```python
#Linear Regression

import matplotlib.pyplot as plt
from scipy import stats

x = data['YearsExperience']
y = data['Salary']

slope , intercept , r , p , std_err = stats.linregress(x,y)

def myfun(x):
    return slope * x + intercept

mymodel = list(map(myfun,x))

plt.scatter(x,y)
plt.plot(x,mymodel)
plt.show()
```

```python
print(r)
```

```python
predicted = myfun(5)

print(predicted)
```

**Task 2: Polynomial Regression**

```python
data1 = pd.read_csv('/content/rank_salary.csv')

plt.figure(figsize=(10, 6))
plt.scatter(data1['Level'], data1['Salary'], color='blue', alpha=0.5)

plt.title('Scatter Diagram')
plt.xlabel('Level')
plt.ylabel('Salary')

plt.grid()
plt.show()
```

```python
import numpy as np

x1 = data1['Level']
y1 = data1['Salary']

mymodel = np.poly1d(np.polyfit(x1,y1,3))
myline = np.linspace(1,22,100)

plt.scatter(x1,y1)
plt.plot(myline,mymodel(myline))
plt.show()
```

```python
from sklearn.metrics import r2_score

mymodel = np.poly1d(np.polyfit(x1,y1,3))

print(r2_score(y1,mymodel(x1)))
```

```python
level = mymodel(5)
print(level)
```

**Task 3: Multiple Regression**

```python
import pandas
from sklearn import linear_model

df2 = pandas.read_csv("/content/multi_regression_data.csv")

x2 = df2[['Weight', 'Volume']]
y2 = df2['CO2']

reg = linear_model.LinearRegression()
reg.fit(x2, y2)

pred = reg.predict([[2500, 1700]])

print(pred)
```

```python
print(reg.coef_)
```

```python
pred1 = reg.predict([[3500, 1700]])

print(pred1)
```

```python
import pandas
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()

df3 = pandas.read_csv("/content/multi_regression_data.csv")

x3 = df3[['Weight', 'Volume']]

scaledX = scale.fit_transform(x3)

print(scaledX)
```

```python
x3 = df3[['Weight', 'Volume']]
y3 = df3['CO2']

scaledX = scale.fit_transform(x3)

reg1 = linear_model.LinearRegression()
reg1.fit(scaledX, y3)

scaled = scale.transform([[2500, 1.7]])

pred2 = reg1.predict([scaled[0]])
print(pred2)
```
